```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 打开文件并把最后的无用字符去了
with open('band_data.dat','r') as f:
    s=f.read().rstrip()

# 按照空行分隔
s_split=s.split('\n\n')
band_number=len(s_split)

# 先建立一个初始的DataFrame
df=pd.DataFrame([s_i.split() for s_i in s_split[0].split('\n')])

'''
每一部分为一个字符串，将一个字符串处理为list
'''
for i in range(1,band_number):
    # 先按行分开，每一部分按照空格分开，换成array形式
    s_i_array=np.array([s_i.split() for s_i in s_split[i].split('\n')])
    # 后面的只取第二列
    df[i]=s_i_array[:,1]
```

```python
# 转为数字并设置index
df=df.astype('float').set_index(0)
```

```python
fig,ax=plt.subplots()
ax.plot(df.index,df.values)
ax.set_xlabel('distance')
ax.set_ylabel('Energy/eV')
ax.set_ylim([-1,1])
ax.set_xticks([df.index[0],df.index[19],df.index[39],df.index[59]])
ax.set_xticklabels(['L',r'$\Gamma$','W','X'])
ax.set_title('band of diamond Si')
```

```python
fig.savefig('band.png',dpi=600)
```
